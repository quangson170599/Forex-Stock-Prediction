```python
import pandas as pd
import time
import os
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('FX_EURUSD, 240 (16).csv')
len(df)
```

```python
cols = ['time','open','high','low','close','EMA','MA','EMA 25','EMA 30','Volume']
df = df[cols]
```

```python
df = df.astype({'time':'datetime64'})
df['date'] = pd.to_datetime(df['time'], unit='s').dt.date
df = df.set_index('date')
df.drop('time',axis=1,inplace=True)
```

```python
df.head()
```

```python
window_size = 10
result = {}
```

```python
feature_columns = df.columns.tolist()
feature_columns
```

```python
len(df)
```

```python
df['future'] = df['close'].shift(-window_size)
last_sequence = df[feature_columns].tail(window_size)
```

```python
df = df.iloc[:-10,:]
```

```python
df['label'] = np.where(df['future']-df['close'] > 0, 1, 0)
```

```python
df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

```python
last_sequence
```

```python
column_scaler = {}
for column in feature_columns:
  scaler = preprocessing.MinMaxScaler()
  new_df[column] = scaler.fit_transform(new_df[column].values.reshape(-1,1))
  column_scaler[column] = scaler
```

```python
for column in feature_columns:
  scaler = preprocessing.MinMaxScaler()
  last_sequence[column] = scaler.fit_transform(last_sequence[column].values.reshape(-1,1))
```

```python
new_df
```

```python
last_sequence
```

```python
new_df.drop('future',inplace=True,axis=1)
```

```python
X = new_df.iloc[:,:-1]
y = new_df.iloc[:,-1]
```

```python
test_size = 0.2
train_samples = int((1 - test_size)*len(X))
```

```python
X_train = X[:train_samples]
y_train = y[:train_samples]
X_test = X[train_samples:]
y_test = y[train_samples:]
```

```python
lgbm_params = {
    'n_estimators': [100,200,300],
    'learning_rate': [0.01, 0.05,0.1,0.3],
    'num_leaves': [20,50,80,100],
    'max_depth' : [3,4,5,6],
    'num_iterations' : [10,20,30,40,50]
}
seed=0
```

```python
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(
    seed=seed,
    objective="binary",
    boosting_type='dart'
)
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(
    estimator=lgbm,
    param_grid=lgbm_params,
    cv=5, scoring='accuracy',
    error_score=0
)
grid_result = clf.fit(X_train, y_train)
```

```python
final_model = lgbm.set_params(**grid_result.best_params_)
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy with LGBM = ',accuracy_score(y_pred, y_test))
```
